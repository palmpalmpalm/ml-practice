```python
import pandas as pd 
data = pd.read_csv('HR_comma_sep.txt')
```

```python
data.head()
```

```python
print(data.Department.value_counts())
```

```python
print(data.salary.value_counts())
```

```python
salary_level = {
    'low' : 1,
    'medium' : 2,
    'high' : 3
}
data['salary'] = data['salary'].map(salary_level)
```

```python
data
```

```python
dummies = pd.get_dummies(data['Department'])
dummies.head()
```

```python
data = pd.concat([data,dummies],axis='columns')
data.head()
```

```python
data = data.drop('Department',axis='columns')
y = data['left']
X = data.drop('left',axis='columns')
```

```python
X.head()
```

```python
y.head()
```

```python
print(X.shape,y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=10000)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.coef_
```

```python
from sklearn.preprocessing import Normalizer
```

```python
scaler = Normalizer()
normalizedX = scaler.fit_transform(X_train)
```

```python
normalizedX
```

```python
model2 = LogisticRegression()
model2.fit(normalizedX,y_train)
```

```python
normalizedX_test = scaler.fit_transform(X_test)
model2.score(normalizedX_test,y_test)
```
